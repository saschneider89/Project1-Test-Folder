```python
import matplotlib
import matplotlib.animation as ani
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os
```

```python
file = ("Hospitalizations_data.csv")
hospitalizations_df = pd.read_csv(file)
hospitalizations_df
```

```python
file2 = ("Total_Hospitalizations_Date")
hospitalization_date_df = pd.read_csv(file2)
hospitalization_date_df
```

```python
hospitalization_date_df
hospitalization_date_df.index = pd.to_datetime(hospitalization_date_df.index)
```

```python
import numpy as np
import matplotlib.pyplot as plt
color = ['red']
fig = plt.figure()
plt.xticks(rotation=45, ha="right", rotation_mode="anchor") #rotate the x-axis values
plt.subplots_adjust(bottom = 0.2, top = 0.9) #ensuring the dates (on the x-axis) fit in the screen
plt.ylabel('Daily Hospitalizations')
plt.xlabel('Dates')

#Curve Function
def buildmebarchart(i=int):
    plt.legend(hospitalization_date_df.columns)
    p = plt.plot(hospitalization_date_df[:i].index, hospitalization_date_df[:i].values) #note it only returns the dataset, up to the point i
    for i in range(0,1):
        p[i].set_color(color[i]) #set the colour of each curve
import matplotlib.animation as ani
animator = ani.FuncAnimation(fig, buildmebarchart, interval = 100)
plt.show()
```
